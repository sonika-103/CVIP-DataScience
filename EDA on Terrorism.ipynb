```python
plt.figure(figsize=(40,20))
sns.countplot('Year',data=Terror)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

```python
plt.figure(figsize=(40,20))
sns.countplot('AttackType',data=Terror, color='Green')
plt.title('Attacking Method by Terrorists')
plt.show()
```

```python
plt.figure(figsize=(40,20))
sns.countplot('AttackType',data=Terror, palette='rocket', order=Terror['AttackType'].value_counts().index)
plt.title('Attacking Method by Terrorists')
plt.show()
```

```python
plt.figure(figsize=(40,20))
sns.countplot('Target_type',data=Terror, order=Terror['Target_type'].value_counts().index)
plt.title('Target Preferences of Terrorists')
plt.show()
```

```python
plt.figure(figsize=(40,20))
sns.countplot('Country',data=Terror, order=Terror['Country'].value_counts().index)
plt.title('Most affected Countries')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=Terror['Country'].value_counts()[:15].index,
            y=Terror['Country'].value_counts()[:15].values,
            palette='inferno')
plt.title('Most affected Countries')
plt.show()
```

```python
plt.figure(figsize=(40,20))
sns.countplot('Region',data=Terror, order=Terror['Region'].value_counts().index)
plt.title('Number of terrorist activities in the region')
plt.show()
```

```python
Terror_region = pd.crosstab(Terror['Year'], Terror['Region'])
Terror_region.plot(kind='line')
plt.title('Number of terrorist activities in the region')
plt.show()
```

```python
coun_Terror=Terror['Country'].value_counts()[:15].to_frame()
coun_Terror.columns=['Attacks']
coun_kill=Terror.groupby('Country')['Killed'].sum().to_frame()
coun_Terror.merge(coun_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```

```python
sns.barplot(Terror['Group'].value_counts()[1:15].values,Terror['Group'].value_counts()[1:15].index,palette=('inferno'))
plt.title('Terrorists Groups')
plt.show()
```

```python
top_groups10=Terror[Terror['Group'].isin(Terror['Group'].value_counts()[1:11].index)]
pd.crosstab(top_groups10.Year,top_groups10.Group).plot(color=sns.color_palette('Paired',10))
fig.set_size_inches(10,6)
plt.show()
```
